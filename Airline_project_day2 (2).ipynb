## 2. DataFrame Creation with Corrupt Record Handling
- `header = true`
- `mode = permissive` — keeps all rows; bad rows go into `_corrupt_record`
- Null values → replaced with empty string
- Empty values → kept as empty string

````python
from pyspark.sql.functions import col, when, lit

required_cols = [
    "FlightDate", "Reporting_Airline", "IATA_CODE_Reporting_Airline",
    "Tail_Number", "Flight_Number_Reporting_Airline", "Origin", "OriginCityName",
    "OriginState", "Dest", "DestCityName", "DestState", "CRSDepTime", "DepTime",
    "DepDelay", "DepDelayMinutes", "DepDel15", "DepartureDelayGroups",
    "CRSArrTime", "ArrTime", "ArrDelay", "ArrDelayMinutes", "ArrDel15",
    "ArrivalDelayGroups", "Cancelled", "CancellationCode", "Diverted",
    "CRSElapsedTime", "ActualElapsedTime", "AirTime", "Flights", "Distance",
    "DistanceGroup", "CarrierDelay", "WeatherDelay", "NASDelay",
    "SecurityDelay", "LateAircraftDelay"
]

# Step 1: Select required columns + corrupt column (if exists)
df = df.select(*required_cols, "_corrupt_record")

# Step 2: Normalize corrupt column
df = df.withColumn(
    "_corrupt_record",
    when(col("_corrupt_record").isNotNull(), col("_corrupt_record")).otherwise(lit(None))
)
````

## 3. Dictionary-Based Null Percentage Calculation
Using a dictionary where:
- **Key** = column name
- **Value** = initial value (0)

Loop through the dictionary, use `f.col(column).isNull()`, convert Null → 1 and Not Null → 0, then calculate:

```
null_percentage = sum(column_null_flag) / total_rows
```

````python
from pyspark.sql import functions as f

# Total rows in the DataFrame
total_rows = df.count()
print("Total Records:", total_rows)

# Dictionary: key = column name, value = initial null count (0)
null_dict = {
    "Origin":            0,
    "Cancelled":         0,
    "Distance":          0,
    "ArrDelay":          0,
    "DepDelay":          0,
    "Reporting_Airline": 0

}

# Loop through the dictionary and calculate null percentage for each column
print("\n  Null Percentage using Dictionary Loop ")
for column, _ in null_dict.items():
    # Use isNull() and convert: Null -> 1, Not Null -> 0
    column_null_flag = f.when(f.col(column).isNull(), 1).otherwise(0)
    
    # Sum the null flags
    null_sum = df.select(f.sum(column_null_flag).alias("null_count")).collect()[0]["null_count"]
    
    # Calculate null percentage
    null_percentage = null_sum / total_rows
    
    print(f"Column: {column} | Null Count: {null_sum} | Null Percentage: {round(null_percentage * 100, 4)}%")
````

## 4. Null Indicator Columns using `withColumn` and `withColumns`

### Approach 1: Using `withColumn()` — one column at a time
Logic:
- If value is null → 1
- Else → 0

````python
from pyspark.sql import functions as f

# Approach 1: withColumn() — adding null indicator columns one by one
df_with_null = df \
    .withColumn(
        "flight_date_null",
        f.when(f.col("FlightDate").isNull(), 1).otherwise(0)
    ) \
    .withColumn(
        "reporting_airline_null",
        f.when(f.col("Reporting_Airline").isNull(), 1).otherwise(0)
    )

print("Approach 1 - withColumn() result:")
df_with_null.select(
    "FlightDate", "flight_date_null",
    "Reporting_Airline", "reporting_airline_null"
)
# .display()
````

### Approach 2: Using Dictionary + `withColumns()` — multiple columns at once
- Keys → column names
- Values → null logic expressions

Purpose: Compare `withColumn()` (single column) vs `withColumns()` (multiple columns using dictionary)

````python
from pyspark.sql import functions as f

# Approach 2: Dictionary + withColumns() — all null indicator columns at once
# Dictionary: key = new null column name, value = null logic expression
null_column_dict = {
    "flight_date_null":       f.when(f.col("FlightDate").isNull(),           1).otherwise(0),
    "reporting_airline_null": f.when(f.col("Reporting_Airline").isNull(),     1).otherwise(0),
    "origin_null":            f.when(f.col("Origin").isNull(),                1).otherwise(0),
    "destination_null":       f.when(f.col("Dest").isNull(),                  1).otherwise(0),
    "distance_null":          f.when(f.col("Distance").isNull(),              1).otherwise(0),
    "arrival_delay_null":     f.when(f.col("ArrDelay").isNull(),             1).otherwise(0),
    "departure_delay_null":   f.when(f.col("DepDelay").isNull(),              1).otherwise(0),
    "cancelled_null":         f.when(f.col("Cancelled").isNull(),             1).otherwise(0),
}

# withColumns() accepts a dictionary directly — much cleaner than chaining withColumn()
df_with_null_cols = df.withColumns(null_column_dict)

print("Approach 2 - withColumns() with dictionary result:")
df_with_null_cols.select(
    "FlightDate",         "flight_date_null",
    "Reporting_Airline",  "reporting_airline_null",
    "Origin",             "origin_null",
    "Dest",               "destination_null"
)
# .display()


# Comparison Summary:
# withColumn()  -> adds ONE column at a time, good for simple cases
# withColumns() -> adds MULTIPLE columns in one call using a dict, cleaner & more efficient
print("\nComparison:")
print("withColumn()  -> Single column addition (chained multiple times)")
print("withColumns() -> Multiple columns at once using a dictionary (more efficient)")
````

## 5. Final Bronze DataFrame Creation
Including:
- `flight_date`, `reporting_airline`, `origin`, `destination`
- All null indicator columns
- Corrupt record column

````python
from pyspark.sql import functions as f

# Final Bronze DataFrame
df_bronze = df_with_null_cols.select(
    # Core identity columns
    f.col("FlightDate"),
    f.col("Reporting_Airline"),
    f.col("Origin"),
    f.col("Dest"),

    # All null indicator columns
    "flight_date_null",
    "reporting_airline_null",
    "origin_null",
    "destination_null",
    "distance_null",
    "arrival_delay_null",
    "departure_delay_null",
    "cancelled_null",

    # Corrupt record column
    f.col("_corrupt_record").alias("corrupt_record"),

    # Extra columns needed for validations below
    f.col("DepDelay"),
    f.col("Distance")
)

print("Final Bronze DataFrame — Total Rows:", df_bronze.count())
print("Columns:", df_bronze.columns)
# df_bronze.display()
````

## 6. Departure Delay Range Validation
- Valid range: **-120 minutes** (2 hrs early) to **1440 minutes** (24 hrs delay)
- If within range → 0 (valid)
- Else → 1 (invalid)
- Creates column: `departure_delay_range_flag`

````python
from pyspark.sql import functions as f

# Valid departure delay range: -120 (2 hours early) to 1440 (24 hours delay)
df_bronze = df_bronze.withColumn(
    "departure_delay_range_flag",
    f.when(
        (f.col("DepDelay") >= -120) & (f.col("DepDelay") <= 1440),
        0  # Valid
    ).otherwise(1)  # Invalid
)

# Show distribution
print("Departure Delay Range Validation (0 = valid, 1 = invalid):")
df_bronze.groupBy("departure_delay_range_flag").count().show()
````

## 7. Same Origin-Destination Check
- Creates column: `same_origin_destination`
- If `origin == destination` → 1 (suspicious/invalid flight)
- Else → 0 (valid)

````python
from pyspark.sql import functions as f

# Same origin-destination check
df_bronze = df_bronze.withColumn(
    "same_origin_destination",
    f.when(f.col("origin") == f.col("dest"), 1).otherwise(0)
)

# Show distribution
print("Same Origin-Destination Check (1 = same/invalid, 0 = different/valid):")
df_bronze.groupBy("same_origin_destination").count().show()
````

## 8. Domestic Flight Distance Validation
- Valid range: **1 km to 10,000 km**
- If within range → 0 (valid)
- Else → 1 (invalid)
- Creates column: `check_distance_range`

````python
from pyspark.sql import functions as f

# Domestic flight distance valid range: 1 to 10,000 km
df_bronze = df_bronze.withColumn(
    "check_distance_range",
    f.when(
        (f.col("Distance") >= 1) & (f.col("Distance") <= 10000),
        0  # Valid
    ).otherwise(1)  # Invalid
)

# Show distribution
print("Distance Range Validation (0 = valid, 1 = invalid):")
df_bronze.groupBy("check_distance_range").count().show()
````
